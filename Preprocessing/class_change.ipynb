```python
from glob import glob
```

```python
labels = ['bench_press_up']

dir = './bench_press/labels/'


for i in range(len(labels)):
    label_list = glob(f'{dir}{labels[i]}*.txt')
    for file in label_list:
        with open(file, 'r') as f:
            lines = f.readlines()
        for j in range(len(lines)):
            rep = lines[j].replace("1", "0", 1)
            lines[j] = rep
        with open(file, 'w') as f:
            f.write("".join(lines)[:-1])
        with open(file, 'r') as f:
            print(f.readline())
```

```python
with open(dir+'classes.txt', 'w') as f:
    for i in range(len(labels)):
      f.write(labels[i])
      if i != (len(labels)-1):
        f.write('\n')

with open('./bench_press/labels/classes.txt', 'r') as f:
    print(f.read())
```
